```python
import numpy as np
import pandas as pd
```

```python
transform_table=pd.read_excel(io="../data/transform_table.xlsx")
```

```python
base_data=pd.read_excel(
    io="../data/Data Landscaping Master (Draft) - PQ WIP.xlsx",
    sheet_name="Draft Master PQ",
    skiprows=6,
    usecols=["S level segment","S level closest match"]
)
```

```python
def segment_mapper(segment:str,transform_table:pd.DataFrame)->str:
    """
    Simple function to get segment mapping.

    Paramter(s):
    - segment, str
    - transform_table, pd.DataFrame

    Returns:
    - mapped_segment, str
    """
    subdf=transform_table[transform_table["From"]==segment]
    mapped_segments=subdf["To"].values
    exact_match=segment in mapped_segments
    if len(mapped_segments)==0:
        return [exact_match,segment]
    elif len(mapped_segments)==1:
        return [exact_match,mapped_segments[0]]
    else:
        return [exact_match,';'.join(mapped_segments)]
```

```python
mapped_segments=pd.DataFrame(
    data=base_data["S level segment"].apply(segment_mapper,args=(transform_table,)).tolist(),
    columns=["exact_match","mapped_segments"]
)
```

```python
results=pd.concat(
    objs=[base_data,mapped_segments],
    axis=1
)
```

```python
results.to_csv(path_or_buf="../results/mapped_segments.csv",index=False)
```
